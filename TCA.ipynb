```python
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'])
ct_percentage = ct.div(ct.sum(axis=1), axis=0) * 100

ax = ct_percentage.plot(kind='bar', stacked=True, figsize=(10, 6), cmap='viridis')

plt.title("Churn by Senior Citizen (Percentage)")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage")

for container in ax.containers:
    for bar in container:
        height = bar.get_height()
        if height > 0:  # Only label non-zero bars
            ax.annotate(f'{height:.1f}%', 
                        (bar.get_x() + bar.get_width() / 2, bar.get_y() + height / 2),
                        ha='center', va='center', fontsize=10, color='white')

plt.legend(title="Churn")
plt.tight_layout()
plt.show()
```

```python
# List of columns to create count plots for
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Number of rows and columns for subplots
num_cols = 3
num_rows = (len(columns) + num_cols - 1) // num_cols  # Calculate the required number of rows

# Create subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10))
axes = axes.flatten()  # Flatten the array to easily iterate

# Plot count plots for each column
for i, column in enumerate(columns):
    sns.countplot(data=df, x=column, ax=axes[i] , hue = "Churn")
    axes[i].set_title(f'Count Plot of {column}')

# Hide any unused subplots
for i in range(len(columns), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```

#Most customers have PhoneService, but fewer use MultipleLines. OnlineSecurity, OnlineBackup, and TechSupport are less common, and their absence correlates with higher churn. StreamingTV shows a more balanced distribution, with churn similar for users and non-users. Overall, lacking support and security services seems linked to higher churn rates.
